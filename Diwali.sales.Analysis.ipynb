# Exploratory Data Analysis

### Gender

```python
ax= sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

##### Note :- in that graph show female purchase more than male

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(data=sales_gen, x='Gender', y='Amount')
plt.show()
```

#### Note: From the above graph we can see that most of the buyers are females and even the purchasing power of female are greater than men


### Age

```python
df.columns
```

```python
sns.countplot(data=df,x='Age Group')
```

```python
ax2=sns.countplot(data=df,x='Age Group', hue='Gender')
for bars in ax2.containers:
    ax2.bar_label(bars)
    
```

```python
# Total Amount vs Age Group
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_age
```

```python
ax3= sns.barplot(data=sales_age, x='Age Group' , y='Amount')
for bars in ax3.containers:
    ax3.bar_label(bars)
```

##### Note:-From above graph we can see that most of the buyers are of age group between 26-35 yrs female

### State

```python
df.columns
```

```python
# Number of order from  10 state
order_state= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=order_state, x='State', y='Orders')
```

```python
# Total amount of sales from 10 state
sales_am=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_am, x='State', y='Amount')
```

##### From the above Graph, we can see that most of the orders are from Uttar Pradesh, Maharashtra, and Karnataka respectively  but total 
sales/Amount is the UP, Karnataka, and the Maharashtra


### Marital_Status

```python
ax=sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales, x='Marital_Status', y='Amount', hue='Gender')
```

#### from above graph we can see that most of the buyers are married women and they purchasing power

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(['Occupation','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales, x='Occupation', y='Amount', hue='Gender')
```

##### From above graph we can see that most of the buyers are working in IT, Aviation and Healthcare sector

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(25,5)})
ax5=sns.countplot(data=df, x='Product_Category')

for bars in ax5.containers:
    ax5.bar_label(bars)
```

```python
sales= df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=sales, x='Product_Category', y='Amount') 
```

##### From above graph we can see that most of the sold products are from food,clothing and electronic category

```python
sales_state =df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='Product_ID', y='Orders')
```

### Conclusion:

The Diwali sales data analysis reveals a significant increase in consumer spending, with electronics and apparel sectors showing the highest growth rates. Online shopping platforms experienced a notable surge, driven by exclusive discounts and the convenience of home delivery. Additionally, regional trends indicated a higher expenditure in urban areas compared to rural regions. The overall positive sales performance underscores the festive season's critical role in boosting economic activity, highlighting the effectiveness of targeted marketing strategies and promotional campaigns. This analysis provides valuable insights for retailers to optimize future sales efforts and capitalize on festive buying behaviors.
